This is interesting. The *age* variable almost looks like it is Normal Distributed, but it is a little skewed to the right. However, the same does not apply to the rest of the variables. Finally, the *hours rates* and *month rates* appear to follow a uniform distribution, which shows that there is almost the same amount of employees in each rate interval.

We will have a closer look to the age variable. We can analyze if it actually follows a Normal Distribution or not. Apart from being something nice to know, it would be useful if we were to make a model with our data and if we had to make assumptions about the distribution of the variable.

```python
employee_df['Age'].hist(figsize=(4,4))
```

```python
age_mean = employee_df['Age'].mean()
age_std = employee_df['Age'].std()

print(age_mean, age_std)
```

We know the mean and the standard deviation of the distribution. Now, in order to check if this ressembles a normal distribution, we have to know if this follows the impirical rule:

1. 68% of values fall within 1 standard deviation of the mean
1. 95% of values fall within 2 standard deviation of the mean 
1. 99.7% of values fall within 3 standard deviation of the mean 

```python
values = {}

for i in range(1,4):
    lower_limit = age_mean - i * age_std
    upper_limit = age_mean + 1 * age_std

    result = ((employee_df['Age'] >= lower_limit) & (employee_df['Age'] <= upper_limit)).mean()

    if i not in values:
        values[i] = [result]
    else: 
        values[i].append(result)

values
```

**Conclusion**: As similar as the distribution may look, the data does not support that the Age follows a Normal Distribution.

### **Data Relationships**
Now that we have explored the data, it is time to analyze what are the relationships that we will visualize in the Looker Studio. The main objective is to study the Attrition of these employees.

```python
employee_df.head()
```

First, we need to see how many employees left the company in general

```python
plt.bar(employee_df['Attrition'].unique(), employee_df['Attrition'].value_counts(), color = ['red', 'green'])
```

Apparently, there are more people inside this dataset that left the company, than people that did not. Now, we will analyze wether there are any strong relationships between Attrition and other employees' characteristics.

```python
# We define a target variable to correlate aginst the rest:

target_variable = 'Attrition'
```

```python
# Then, we have to re-map the existing variables to their corresponding values. For example, Education or Environment Satisfaction

# Maps
mapping_education = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
mapping_env_satisfaction = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
mapping_job_involvement = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
mapping_job_satisfaction = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
mapping_performance = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
mapping_relationship = {1: 'Low', 2: 'Good', 3: 'High', 4: 'Very High'}
mapping_work_balance = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}

# Re-mapping
employee_df['Education'] = employee_df['Education'].map(mapping_education)
employee_df['EnvironmentSatisfaction'] = employee_df['EnvironmentSatisfaction'].map(mapping_env_satisfaction)
employee_df['JobInvolvement'] = employee_df['JobInvolvement'].map(mapping_job_involvement)
employee_df['JobSatisfaction'] = employee_df['JobSatisfaction'].map(mapping_job_satisfaction)
employee_df['PerformanceRating'] = employee_df['PerformanceRating'].map(mapping_performance)
employee_df['RelationshipSatisfaction'] = employee_df['RelationshipSatisfaction'].map(mapping_relationship)
employee_df['WorkLifeBalance'] = employee_df['WorkLifeBalance'].map(mapping_work_balance)
```
